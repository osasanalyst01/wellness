## Analyse regional sales data to identify strong and weak markets. You can use geo-visualizations at this point.

There is no column for longitude and latitude so we are going to generate random longitude and latitude that corresponds
#to North, East, West and South of United states to visualize our data better

```python
# Function to generate random coordinates
def generate_coordinates(lat_range, lon_range, n):
    lat = np.random.uniform(lat_range[0], lat_range[1], n)
    lon = np.random.uniform(lon_range[0], lon_range[1], n)
    return lat, lon

# Generate coordinates for North region
north_lat, north_lon = generate_coordinates((45, 49), (-124, -67), len(df1[df1['Region'] == 'North']))

# Generate coordinates for East region
east_lat, east_lon = generate_coordinates((34, 45), (-67, -80), len(df1[df1['Region'] == 'East']))

# Generate coordinates for West region
west_lat, west_lon = generate_coordinates((34, 45), (-124, -117), len(df1[df1['Region'] == 'West']))

# Generate coordinates for South region
south_lat, south_lon = generate_coordinates((25, 34), (-124, -67), len(df1[df1['Region'] == 'South']))

# Assign coordinates to corresponding rows in dataframe
df1.loc[df1['Region'] == 'North', ['Latitude', 'Longitude']] = list(zip(north_lat, north_lon))
df1.loc[df1['Region'] == 'East', ['Latitude', 'Longitude']] = list(zip(east_lat, east_lon))
df1.loc[df1['Region'] == 'West', ['Latitude', 'Longitude']] = list(zip(west_lat, west_lon))
df1.loc[df1['Region'] == 'South', ['Latitude', 'Longitude']] = list(zip(south_lat, south_lon))

df1.head(10)
```

```python
# To create a geeo-visualization for sales per region
fig = go.Figure()

# Add sales amount per point
fig.add_trace(go.Scattergeo(
    lon=df1['Longitude'],
    lat=df1['Latitude'],
    text=df1['Region'] + '<br>Sales: $' + df1['Sales_Amount'].astype(str),
    mode='markers',
    marker=dict(
        size=df1['Sales_Amount'] / 50,  
        color=df1['Sales_Amount'],  
        colorscale='YlOrRd', 
        showscale=True,  
        colorbar_title='Sales Amount'  
    )
))

# Create a dictionary of region names and their respective center coordinates to place total sales for each region
region_coordinates = {
    'North': {'Latitude': 48.27458, 'Longitude': -101.4648},
    'East': {'Latitude': 39.51864, 'Longitude': -79.68222},
    'West': {'Latitude': 38.88764, 'Longitude': -123.039},
    'South': {'Latitude': 28.56189, 'Longitude': -98.6311}
}

# Calculate the sum of sales amount for each region
region_totals = df1.groupby('Region')['Sales_Amount'].sum().reset_index()
region_totals = region_totals.merge(pd.DataFrame(region_coordinates).T,
                                    left_on='Region', right_index=True)
color_scale = [(0, 'yellow'), (1, 'red')]

# Add total sales amount for each region as annotations
for region, coord in region_coordinates.items():
    total_sales = region_totals[region_totals['Region'] == region]['Sales_Amount'].values[0]
    fig.add_trace(go.Scattergeo(
        lon=[coord['Longitude']],
        lat=[coord['Latitude']],
        text=f"Total Sales: ${total_sales}",
        mode='markers+text',
        marker=dict(
            size=total_sales / 1000,  
            color=total_sales, 
            colorscale=color_scale,  
            opacity=0.5,  
            line=dict(color='black', width=2),  
            symbol='circle',  
            cmin=region_totals['Sales_Amount'].min(),  
            cmax=region_totals['Sales_Amount'].max(),  
            colorbar=dict(title='Total Sales')  
        ),
        textfont=dict(size=12, color='black'), 
        textposition='bottom center',  
        showlegend=False  
    ))


fig.update_layout(
    title='Sales Distribution per Region',
    geo=dict(
        scope='usa',  
        projection_type='albers usa' 
    )
)

# Show plot
fig.show()
```

```python
#To create geo-visualization of transaction count per region

fig = go.Figure()

# Add transaction count per point
fig.add_trace(go.Scattergeo(
    lon=df1['Longitude'],
    lat=df1['Latitude'],
    text=df1['Region'] + '<br>Transactions: ' + df1['Transaction_Count'].astype(str),
    mode='markers',
    marker=dict(
        size=df1['Transaction_Count'] / 30,  
        color=df1['Transaction_Count'],  
        colorscale='YlOrRd',  
        showscale=True,  
        colorbar_title='Transaction Count'  
    )
))

# Create a dictionary of region names and their respective center coordinates
region_coordinates = {
    'North': {'Latitude': 48.27458, 'Longitude': -101.4648},
    'East': {'Latitude': 39.51864, 'Longitude': -79.68222},
    'West': {'Latitude': 38.88764, 'Longitude': -123.039},
    'South': {'Latitude': 28.56189, 'Longitude': -98.6311}
}

# Calculate the sum of transaction count for each region
region_totals = df1.groupby('Region')['Transaction_Count'].sum().reset_index()
region_totals = region_totals.merge(pd.DataFrame(region_coordinates).T,
                                    left_on='Region', right_index=True)

# Add total transaction count for each region as annotations
for region, coord in region_coordinates.items():
    total_transactions = region_totals[region_totals['Region'] == region]['Transaction_Count'].values[0]
    fig.add_trace(go.Scattergeo(
        lon=[coord['Longitude']],
        lat=[coord['Latitude']],
        text=f"Total Transactions: {total_transactions}",
        mode='markers+text',
        marker=dict(
            size=total_transactions / 100,  
            color=total_transactions,  
            colorscale='YlOrRd', 
            opacity=0.5,  
            line=dict(color='black', width=2),  
            symbol='circle',  
            cmin=region_totals['Transaction_Count'].min(),  
            cmax=region_totals['Transaction_Count'].max(),  
            colorbar=dict(title='Total Transactions')  
        ),
        textfont=dict(size=12, color='black'),  
        textposition='bottom center',  
        showlegend=False 
    ))


fig.update_layout(
    title='Transaction Count Distribution per Region',
    geo=dict(
        scope='usa',  
        projection_type='albers usa'  
    )
)

# Show plot
fig.show()
```
